```python
Location = r'C:\Users\l1nsb01\Desktop\Data Science Class\Capstone\Dual_Stock_C.xlsx'
df = pd.read_excel(Location) # Load data from spreadsheet
```

```python
df.head() #view document columns in header
```

```python
# Create a new column that shows percent utilization for shanghai southbound stock channel
df['SSE_South_Util'] = 10500000 - df['H1DQB']
df['SSE_South_Util'] = df['SSE_South_Util'] / 10500000
df.head()
```

```python
# Create a new column that shows percent utilization for shanghai northbound stock channel
df['SSE_North_Util'] = 13000000 - df['C1DQB']
df['SSE_North_Util'] = df['SSE_North_Util'] / 13000000
df.head()
```

```python
# Create column showing the premium of offshore renminib relative to onshore renminbi
df['FX_Prem'] = df['HKDCNH'] / df['HKDCNY']
df.head()
```

```python
# Rename columns for better readability
df = df.rename(columns={
    'H1DQB': 'SSE_South',
    'C1DQB': 'SSE_North',
    'H2DQB': 'SZSE_South',
    'C2DQB': 'SZSE_North',
    'HSAHP': 'AH_Premium',
    })
df.head()
```

```python
# Chart showing the A-H Share Premium over time
plt.plot(df.Date, df.AH_Premium)
plt.xlabel("Date")
plt.ylabel("Premium")
```

```python
# Chart showing Shanghai Southbound Utilization over time
plt.plot(df.Date, df.SSE_South_Util)
plt.xlabel("Date")
plt.ylabel("Utilization")
```

```python
# Chart showing Shanghai northbound Utilization over time
plt.plot(df.Date, df.SSE_North_Util)
plt.xlabel("Date")
plt.ylabel("Utilization")
```

```python
# Chart showing Offshore/Onshore Renminbi premium
plt.plot(df.Date, df.FX_Prem)
plt.xlabel("Date")
plt.ylabel("Premium")
```
